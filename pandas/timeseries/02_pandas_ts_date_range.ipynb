<h1 style="color:blue;" align="center">Pandas Time Series Analysis Tutorial: date_range</h1>

```python
import pandas as pd

df = pd.read_csv('aapl_no_dates.csv')
df.head()
```

```python
# slect the data based on some data range
# freq='B' stands for select data of business date(mom-friday)
rng = pd.date_range(start='6/1/2017', end='6/30/2017', freq='B')
rng
```

```python
# set the business data as index
df.set_index(rng, inplace=True)
df
```

```python
# plot the 'close' column of dataframe
%matplotlib inline
df.Close.plot()
```

```python
df.Close.plot(kind='bar')
```

```python
# select rows based on some date range
# find the mean 
df["2017-06-01":"2017-06-10"].Close.mean()
```

```python
# select all the days
# as non business dates will not have any stock price data
# the pad method will be used to keep the previous date records in non business days
df.asfreq('D', method='pad')
```

```python
df.asfreq('W', method='pad')
```

```python
df.asfreq('H', method='pad')
```

```python
# slect the data based on some data range
# freq='B' stands for select data of business date(mom-friday)
# insted of end date we can also use no of periods 
# will generate 72 days exclude non business dates
# this is helpful when we don't know the end date insted we want to see the data for some periods or next n days(working days)
rng = pd.date_range(start='6/1/2017', periods=72, freq='B')
rng
```

```python
import numpy as np
np.random.randint(1,10,len(rng)) # will give 72 random generated number of range(1,10)
```

```python
# create a time series  
ts = pd.Series(np.random.randint(1,10,len(rng)), index=rng)
ts
```
